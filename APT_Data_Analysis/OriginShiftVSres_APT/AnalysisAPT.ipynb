```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp_var = pd.read_csv("OriginVSresAPT.csv")
DATA = np.transpose(temp_var.to_numpy())
DATAp1 = DATA[0][DATA[1] == 0.1 ], DATA[2][DATA[1] == 0.1 ]
DATAp15 = DATA[0][DATA[1] == 0.15 ], DATA[2][DATA[1] == 0.15 ]
DATAp2 = DATA[0][DATA[1] == 0.2 ], DATA[2][DATA[1] == 0.2 ]
DATAp3 = DATA[0][DATA[1] == 0.3 ], DATA[2][DATA[1] == 0.3 ]
DATAp5 = DATA[0][DATA[1] == 0.5 ], DATA[2][DATA[1] == 0.5 ]
DATA1 = DATA[0][DATA[1] == 1 ], DATA[2][DATA[1] == 1 ]
DATA2 = DATA[0][DATA[1] == 2 ], DATA[2][DATA[1] == 2 ]
```

```python
plt.figure(figsize=(16,18))
plt.suptitle("Avg active cell count v/s $ios$ (in nanometers)", fontsize=18)

plt.subplot(4,2,1)
plt.plot(DATAp1[0],DATAp1[1],'sb--')
plt.ylabel("avg. active cell count")
plt.title("RES = 0.1")

plt.subplot(4,2,2)
plt.plot(DATAp15[0],DATAp15[1],'Dr--')
plt.ylabel("avg. active cell count")
plt.title("RES = 0.15")

plt.subplot(4,2,3)
plt.plot(DATAp2[0],DATAp2[1], 'ok--')
plt.ylabel("avg. active cell count")
plt.title("RES = 0.2")

plt.subplot(4,2,4)
plt.plot(DATAp3[0],DATAp3[1],'^m--')
plt.ylabel("avg. active cell count")
plt.title("RES = 0.3")

plt.subplot(4,2,5)
plt.plot(DATAp5[0],DATAp5[1], '+c--')
plt.ylabel("avg. active cell count")
plt.title("RES = 0.5")

plt.subplot(4,2,6)
plt.plot(DATA1[0],DATA1[1],'*y--')
plt.xlabel("origin shift $ios$")
plt.ylabel("avg. active cell count")
plt.title("RES = 1")

plt.subplot(4,2,7)
plt.plot(DATA2[0],DATA2[1],'xg--')
plt.xlabel("origin shift $ios$")
plt.ylabel("avg. active cell count")
plt.title("RES = 2")

plt.savefig("aaccVSios.png")
```

```python
RESs = [0.1, 0.15, 0.2, 0.3, 0.5, 1,2]
for i in RESs :
    data = DATA[2][DATA[1]==i]
    print(i, ": ", np.average(data))
```
