```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tempIDEAL = pd.read_csv("Origin_shiftVSresolution.csv")
tempAPT = pd.read_csv("../OriginShiftVSres_APT/OriginVSresAPT.csv")
DATAIDEAL = np.transpose(tempIDEAL.to_numpy())
DATAAPT = np.transpose(tempAPT.to_numpy())
```

```python
DATAAPT[0] = DATAAPT[0]*10
DATAAPT[1] = DATAAPT[1]*10
```

All units are expressed in angstroms from here on

```python
np.unique(DATAAPT[1]),np.unique(DATAIDEAL[1])
```

```python
DATA = DATAAPT
APT1 =  DATA[0][DATA[1] == 1 ], DATA[2][DATA[1] == 1 ]
APT1p5 =  DATA[0][DATA[1] == 1.5 ], DATA[2][DATA[1] == 1.5 ]
APT2 =  DATA[0][DATA[1] == 2 ], DATA[2][DATA[1] == 2 ]
APT3 =  DATA[0][DATA[1] == 3 ], DATA[2][DATA[1] == 3 ]
APT5 =  DATA[0][DATA[1] == 5 ], DATA[2][DATA[1] == 5 ]
APT10 =  DATA[0][DATA[1] == 10 ], DATA[2][DATA[1] == 10 ]
APT20 =  DATA[0][DATA[1] == 20 ], DATA[2][DATA[1] == 20 ]

DATA = DATAIDEAL
IDEAL1p5 =  DATA[0][DATA[1] == 1.5 ], DATA[2][DATA[1] == 1.5 ]
IDEAL1p75 =  DATA[0][DATA[1] == 1.75 ], DATA[2][DATA[1] == 1.75 ]
IDEAL2 =  DATA[0][DATA[1] == 2 ], DATA[2][DATA[1] == 2 ]
IDEAL3 =  DATA[0][DATA[1] == 3 ], DATA[2][DATA[1] == 3 ]
IDEAL5 =  DATA[0][DATA[1] == 5 ], DATA[2][DATA[1] == 5 ]
IDEAL10 =  DATA[0][DATA[1] == 10 ], DATA[2][DATA[1] == 10 ]
IDEAL20 =  DATA[0][DATA[1] == 20 ], DATA[2][DATA[1] == 20 ]
```

```python
plt.figure(figsize=(8,6))
APT = APT1p5
IDEAL = IDEAL1p5
plt.plot(APT[0],APT[1],'sb--')
plt.plot(IDEAL[0],IDEAL[1],'^r--')
plt.ylabel("avg. active cell count")
plt.xlabel("origin shift in angstroms")
plt.grid()
plt.title("RES = 1.5")
plt.show()
```
