```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp_var = pd.read_csv("Origin_shiftVSresolution.csv")
DATA = np.transpose(temp_var.to_numpy())
DATA1p5 = DATA[0][DATA[1] == 1.5 ], DATA[2][DATA[1] == 1.5 ]
DATA1p75 = DATA[0][DATA[1] == 1.75 ], DATA[2][DATA[1] == 1.75 ]
DATA2 = DATA[0][DATA[1] == 2 ], DATA[2][DATA[1] == 2 ]
DATA3 = DATA[0][DATA[1] == 3 ], DATA[2][DATA[1] == 3 ]
DATA5 = DATA[0][DATA[1] == 5 ], DATA[2][DATA[1] == 5 ]
DATA10 = DATA[0][DATA[1] == 10 ], DATA[2][DATA[1] == 10 ]
DATA20 = DATA[0][DATA[1] == 20 ], DATA[2][DATA[1] == 20 ]
```

```python
RESs = [1.5, 1.75, 2, 3, 5, 10, 20]
for i in RESs :
    data = DATA[2][DATA[1]==i]
    print(i, ": ", np.average(data))
```

```python
plt.figure(figsize=(8,6))
plt.plot(DATA20[0],DATA20[1],'sb--')
plt.ylabel("avg. active cell count")
plt.grid()
plt.title("RES = 20")
```

```python
plt.figure(figsize=(12,12))
plt.suptitle("Avg active cell count v/s $ios$", fontsize=18)

plt.subplot(3,2,1)
plt.plot(DATA1p5[0],DATA1p5[1],'sb--')
plt.ylabel("avg. active cell count")
plt.title("RES = 1.5")
plt.subplot(3,2,2)

plt.plot(DATA1p75[0],DATA1p75[1],'Dr--')
plt.ylabel("avg. active cell count")
plt.title("RES = 1.75")

plt.subplot(3,2,3)
plt.plot(DATA2[0],DATA2[1], 'ok--')
plt.ylabel("avg. active cell count")
plt.title("RES = 2")

plt.subplot(3,2,4)
plt.plot(DATA3[0],DATA3[1],'^m--')
plt.ylabel("avg. active cell count")
plt.title("RES = 3")

plt.subplot(3,2,5)
plt.plot(DATA5[0],DATA5[1], '+c--')
plt.xlabel("origin shift $ios$")
plt.ylabel("avg. active cell count")
plt.title("RES = 5")

plt.subplot(3,2,6)
plt.plot(DATA10[0],DATA10[1],'*y--')
plt.xlabel("origin shift $ios$")
plt.ylabel("avg. active cell count")
plt.title("RES = 10")

plt.savefig("aaccVSios.png")
```
