```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
```

```python
S = pd.read_csv("mass_spectrum.csv",delimiter=',')
```

```python
spectrum = S.to_numpy()
spectrum_trans = np.transpose(spectrum)
```

```python
plt.figure(figsize=(18,6))
plt.plot(spectrum_trans[0],np.log(spectrum_trans[1]))
plt.show()
```

```python
peak_indices = scipy.signal.find_peaks(spectrum_trans[1], threshold=1e2)
peaks = np.transpose([spectrum[i] for i in peak_indices[0]])
print("Num peaks: ",np.shape(peaks)[1])
peak_ratios = peaks[0]
print("Ratios: ",peak_ratios)
peak_counts = peaks[1]
print("Counts: ", peak_counts)
```

```python
plt.figure(figsize=(18,6))
plt.plot(spectrum_trans[0],np.log(spectrum_trans[1]))
plt.plot(peaks[0], np.log(peaks[1]),'r^')
plt.xlabel("mass-to-charge ratio")
plt.ylabel("Count in log scale")
plt.xlim(0, np.max(peak_ratios)+1)
plt.show()
```
