```python
var('T')
G_FCC = -1000 - 2*T
t1 = 912 ; t2 = 1394 ; t3 = 1538;
y1 = G_FCC.subs(T=t1) ; y2 = G_FCC.subs(T=t2)
var('A,B')
eq1 = y1 == A - B*(t1)^2 ; eq2 = y2 == A - B*(t2)^2
UNK = solve([eq1,eq2],A,B)
G_BCC = A.subs(UNK[0][0]) - B.subs(UNK[0][1])*T^2
G_liq = G_BCC.subs(T=t3) -4*(T-t3)
```

# Q2 Ans:

The internal energy of $BCC$ and $FCC$ iron at room temperature in $eV/unitcell$ are :

```python
U_BCC = -0.4169 ; U_FCC = -0.316
```

The lattice parameters in $\overset{\circ}{A}$ for $BCC$ and $FCC$ iron are:

```python
LP_BCC = 2.84005 ; LP_FCC = 3.6457
```

No of atoms per unit cell in $BCC$ and $FCC$ iron are:

```python
n_BCC = 2 ; n_FCC = 4
```

We know $$\Delta G = U + pV - TS$$
To compare the difference in contribution to $\Delta G$ we should convert them to same units.

$1eV=1.6022*10^{-19} J$ and $1m^3atm = 10^5 J$ and $1\overset{\circ}{A} = 10^{-10}m$

The Internal Energy values in $J$ per atom are : 

```python
U_BCC_atom = (U_BCC/n_BCC)* 1.6022*10^(-19)
U_FCC_atom = (U_FCC/n_FCC)* 1.6022*10^(-19)
print(U_BCC_atom,"J:BCC\t",U_FCC_atom,"J:FCC")
```

$pV$ contribution to $\Delta G$ in $J/atom$ are : 
$$1m^3atm = 10^{5}J$$

```python
pV_BCC = ( 1*( LP_BCC*10^(-10) )^3 )* (10^5) /n_BCC
pV_FCC = ( 1*( LP_FCC*10^(-10) )^3 )* (10^5) /n_FCC
print(pV_BCC,"J:BCC\t",pV_FCC,"J:FCC")
```

The contribution of difference in Internal Energy in $J/atom$ is : 

```python
U_BCC_atom - U_FCC_atom
```

The contribution of difference in $pV$ term in $J/atom$ is : 

```python
pV_BCC - pV_FCC
```

```python
(U_BCC_atom - U_FCC_atom)/(pV_BCC - pV_FCC)
```

The $\Delta U$ term contributes $5$ orders of magnitude more than the $pV$ term to $\Delta G$

# Q3 (b) Ans:

Let the mole fraction of $\alpha,\beta,\gamma$ be $a,b,c$ respectively

```python
var('a,b,c')
```

By the law of conservation of mass and given that the alloy on total contains:
$$40 \% A ; 30\% B ; 30\% C$$
We get three linear equations in 3 variables:

```python
eq1 = 80*a + 10*b + 10*c ==40
eq2 = 10*a + 70*b + 20*c ==30
eq3 = 10*a + 20*b + 70*c ==30
```

```python
solve([eq1,eq2,eq3],a,b,c)
```

Solving the equations give us the mole fractions of the phases as:
$$\alpha = 0.42 $$
$$\beta = 0.29 $$
$$\gamma = 0.29 $$
